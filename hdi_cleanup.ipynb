```python
import pandas as pd
```

```python
hdi_df = pd.read_csv('./datasets/HDI.csv')
hdi_df.head()
```

```python
# Drop columns containing 'Human' in their name
hdi_df = hdi_df.drop(columns=[col for col in hdi_df.columns if 'Human' in col])
hdi_df.head()
```

```python
# Create separate dataframes for expected and mean years of schooling
expected_schooling_df = hdi_df[[col for col in hdi_df.columns if 'Expected Years' in col or 'Country' in col]]
mean_schooling_df = hdi_df[[col for col in hdi_df.columns if 'Mean Years' in col or 'Country' in col]]
expected_schooling_df.head()
mean_schooling_df.head()
```

```python
# Export both dataframes to CSV files in the lstm_datasets directory
expected_schooling_df.to_csv('./datasets/expected_schooling.csv', index=False)
mean_schooling_df.to_csv('./datasets/mean_schooling.csv', index=False)
```
