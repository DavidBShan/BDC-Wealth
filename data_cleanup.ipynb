```python
import pandas as pd
```

```python
file_gdp = "GDP.csv"
df_gdp = pd.read_csv(file_gdp)
print("GDP DataFrame columns:", df_gdp.columns)
if 'code' in df_gdp.columns:
    df_gdp = df_gdp.drop(columns=['code'])
df_gdp.rename(columns={"entity": "Country"}, inplace=True)
df_gdp.replace("..", pd.NA, inplace=True)
df_gdp.head()


```

```python
file_gini = "Gini.csv"
df_gini=pd.read_csv(file_gini)
df_gini = df_gini.drop(columns=["Code"])
df_gini.rename(columns={"Entity": "Country"}, inplace=True)
df_gini
```

```python
file_hdi = "HDI.csv"
df_hdi = pd.read_csv(file_hdi)
df_hdi.replace("NaN", pd.NA, inplace=True)
df_hdi.head()
```

```python
file_infra = "Infrastructure.csv"
df_infra = pd.read_csv(file_infra)
df_infra.head()
```

```python
file_net = "Internet.csv"
df_net = pd.read_csv(file_net)
df_net = df_net.drop(columns=["code"])
df_net.rename(columns={"entity": "Country"}, inplace=True)
df_net.head()
```

```python
file_lit = "Literacy.csv"
df_lit = pd.read_csv(file_lit)
df_lit = df_lit.drop(columns=["Code"])
df_lit.rename(columns={"Entity": "Country"}, inplace=True)
df_lit
```

```python
# Merge multiple DataFrames step by step
combined_df = pd.merge(df_gini, df_gdp, how="outer", on="Country")
combined_df = pd.merge(combined_df, df_hdi, how="outer", on="Country")
combined_df = pd.merge(combined_df, df_infra, how="outer", on="Country")
combined_df = pd.merge(combined_df, df_lit, how="outer", on="Country")
combined_df = pd.merge(combined_df, df_net, how="outer", on="Country")

# Display the resulting DataFrame
print(combined_df)
```

```python
df_gini.rename(columns={"entity": "Entity"}, inplace=True)
df_gdp.rename(columns={"entity": "Entity"}, inplace=True)
df_gdp.rename(columns={"code": "Code"}, inplace=True)
df_gini.rename(columns={"code": "Code"}, inplace=True)
df_gini.replace("..", pd.NA, inplace=True)
df_gdp.replace("..", pd.NA, inplace=True)
df_gdp.head()
```

```python
combined_df = pd.merge(df_gini, df_gdp, how="inner", on=["Entity", "Code"])

# Display the combined dataset
print(combined_df)

# Save to a new CSV file
combined_df.to_csv("combined_dataset.csv", index=False)
```
