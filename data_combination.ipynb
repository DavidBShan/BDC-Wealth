```python
import pandas as pd
```

```python
file_gdp = "./datasets/GDP.csv"
df_gdp = pd.read_csv(file_gdp)
print("GDP DataFrame columns:", df_gdp.columns)
if 'code' in df_gdp.columns:
    df_gdp = df_gdp.drop(columns=['code'])
df_gdp.rename(columns={"entity": "Country"}, inplace=True)
df_gdp.replace("..", pd.NA, inplace=True)
df_gdp.head()


```

```python
file_gini = "./datasets/Gini.csv"
df_gini=pd.read_csv(file_gini)
df_gini = df_gini.drop(columns=["Code"])
df_gini.rename(columns={"Entity": "Country"}, inplace=True)
df_gini
```

```python
file_hdi = "./datasets/HDI.csv"
df_hdi = pd.read_csv(file_hdi)
df_hdi.replace("NaN", pd.NA, inplace=True)
df_hdi.head()
```

```python
file_infra = "./datasets/Infrastructure.csv"
df_infra = pd.read_csv(file_infra)
df_infra.head()
```

```python
file_net = "./datasets/Internet.csv"
df_net = pd.read_csv(file_net)
df_net = df_net.drop(columns=["code"])
df_net.rename(columns={"entity": "Country"}, inplace=True)
df_net.replace("..", pd.NA, inplace=True)
df_net.head()
```

```python
file_lit = "./datasets/Literacy.csv"
df_lit = pd.read_csv(file_lit)
df_lit = df_lit.drop(columns=["Code"])
df_lit.rename(columns={"Entity": "Country"}, inplace=True)
df_lit
```

```python
file_co2 = "./datasets/CO2.csv"
df_co2 = pd.read_csv(file_co2)
df_co2 = df_co2.drop(columns=["Unnamed: 0"])
df_co2.rename(columns={"country": "Country"}, inplace=True)
df_co2.head()
```

```python
file_co22 = "./datasets/CO2#2.csv"
df_co22 = pd.read_csv(file_co22)
df_co22 = df_co22.drop(columns=["Unnamed: 0"])
df_co22.rename(columns={"country": "Country"}, inplace=True)
df_co22.head()
```

```python
file_energy = "./datasets/Energy.csv"
df_energy = pd.read_csv(file_energy)
df_energy.replace("..", pd.NA, inplace=True)
df_energy = df_energy.drop(columns=["YR2022", "YR2023"])
df_energy.rename(columns={"Country Name": "Country"}, inplace=True)
year_columns = df_energy.columns[1:]
rename_dict = {col: col.replace('YR', '') + ' Energy' for col in year_columns}

df_energy.rename(columns=rename_dict, inplace=True)
df_energy.head()
```

```python
file_mortality = "./datasets/Mortality.csv"
df_mortality = pd.read_csv(file_mortality)
year_columns = df_mortality.columns[1:]
rename_dict = {str(year): f'{year} Mortality' for year in range(1990, 2023)}
df_mortality.rename(columns=rename_dict, inplace=True)
df_mortality.head()
```

```python
file_pm10 = "./datasets/PM10.csv"
df_pm10 = pd.read_csv(file_pm10)
df_pm10.rename(columns={"WHO Country Name": "Country"}, inplace=True)
rename_dict = {str(year): f'{year} PM10' for year in range(2010, 2022)}
df_pm10.rename(columns=rename_dict, inplace=True)
df_pm10 
```

```python
# Merge multiple DataFrames step by step
combined_df_wo_infra = pd.merge(df_gini, df_gdp, how="inner", on="Country")
combined_df_wo_infra = pd.merge(combined_df_wo_infra, df_hdi, how="inner", on="Country")
combined_df_wo_infra = pd.merge(combined_df_wo_infra, df_lit, how="inner", on="Country")
combined_df_wo_infra = pd.merge(combined_df_wo_infra, df_net, how="inner", on="Country")
combined_df_wo_infra = pd.merge(combined_df_wo_infra, df_co2, how="inner", on="Country")
combined_df_wo_infra = pd.merge(combined_df_wo_infra, df_co22, how="inner", on="Country")
combined_df_wo_infra = pd.merge(combined_df_wo_infra, df_energy, how="inner", on="Country")
combined_df_wo_infra = pd.merge(combined_df_wo_infra, df_mortality, how="inner", on="Country")

# Display the resulting DataFrame
combined_df_wo_infra
```

```python
combined_df_wo_infra.to_csv("./datasets/combined_df_wo_infra.csv", index=False)
```

```python

combined_df = pd.merge(combined_df_wo_infra, df_infra, how="inner", on="Country")
combined_df
```

```python


# Export to CSV
combined_df.to_csv("./datasets/combined_df.csv", index=False)
```
